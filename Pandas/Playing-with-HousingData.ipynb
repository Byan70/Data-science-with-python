## Working on Missing values

```python
# Median of CRIM Column

median_CRIM = hd['CRIM'].median()

# Filling the 'CRIM' null values with its median value
hd['CRIM'].fillna(median_CRIM, inplace=True)

# Filling null values with median
median_INDUS = hd['INDUS'].median()
hd['INDUS'].fillna(median_INDUS, inplace=True)


# Median or 'ZN'
median_ZN = hd['ZN'].median()

# Filling the 'ZN' null values with its median value
hd['ZN'].fillna(median_ZN, inplace= True)

# Filling null with mode
mode_CHAS = hd['CHAS'].mode()[0]
hd['CHAS'].fillna(mode_CHAS, inplace=True)

# Filling null with median value
median_AGE = hd['AGE'].median()
hd['AGE'].fillna(median_AGE, inplace=True)

# filling null with median value
median_LSTAT = hd['LSTAT'].median()
hd['LSTAT'].fillna(median_LSTAT, inplace=True)
```
