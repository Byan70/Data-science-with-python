## Matplotlib Libraries

<> xlabel() -> sets label for x -axis

<> ylabel() -> Sets label for y -axis

<> title() -> Sets the title for the plot

<> loc -> Is a parameter used in title() to position the title: left, right and center

```python
xpoint = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])

ypoint = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(xpoint, ypoint)

plt.title('Sports Watch Data')

plt.xlabel('Average Pulse') # sets the label for x axis

plt.ylabel('Calorie Burnage') # sets the label for y axis
plt.show()
```

### fontdict

--> fontdict parameter is used for setting Font Properties in title(), xlabel(), ylabel() 

```python

xpoint = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])

ypoint = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family': 'serif', 'color': 'blue', 'size': 15}

font2 = {'family': 'serif', 'color': 'darkred', 'size': 20}

plt.plot(xpoint, ypoint)

plt.title('Sports Watch Data', fontdict = font1)

plt.xlabel('Average Pulse', fontdict = font2) # sets the label for x axis

plt.ylabel('Calorie Burnage', fontdict = font2) # sets the label for y axis
plt.show()
```

## Matplotlib Grid


--> grid() - is used to add grid lines to the plot.


```python

xpoint = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])

ypoint = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family': 'serif', 'color': 'blue', 'size': 15}

font2 = {'family': 'serif', 'color': 'darkred', 'size': 20}

plt.plot(xpoint, ypoint)

plt.grid()

plt.title('Sports Watch Data', fontdict = font1)

plt.xlabel('Average Pulse', fontdict = font2) # sets the label for x axis

plt.ylabel('Calorie Burnage', fontdict = font2) # sets the label for y axis
plt.show()
```

 -> You can use the 'axis' parameter to specify which grid lines to display.
 
 -> Legal values are: 'x', 'y' and 'both'. Default is 'both'

```python
xpoint = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])

ypoint = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family': 'serif', 'color': 'blue', 'size': 15}

font2 = {'family': 'serif', 'color': 'darkred', 'size': 20}

plt.plot(xpoint, ypoint)

plt.grid(axis = 'x', color = 'green',  lw = '0.5', ls = 'dotted')

plt.title('Sports Watch Data', fontdict = font1)

plt.xlabel('Average Pulse', fontdict = font2) # sets the label for x axis

plt.ylabel('Calorie Burnage', fontdict = font2) # sets the label for y axis
plt.show()
```


## Matplotlib subplot



-> With the subplot() function you can draw multiple plots in one figure.

-> The layour is organized in rows and columns, which are represented by the first and second argument.

-> The third argument represents the index of the current plot.


```python
# plot 1:

sub1 = np.array([0, 1, 2, 3])

sub2 = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)

plt.plot(sub1, sub2)

# plot 2:

sub3 = np.array([0, 1, 2, 3])

sub4 = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)

plt.plot(sub3, sub4)
plt.show()
```


-> So if we want a figure with 2 rows and 1 column(meaning that the two plots will be displayed on top of
of each other instead of side-by-side)


```python
# plot 1:

sub1 = np.array([0, 1, 2, 3])

sub2 = np.array([3, 8, 1, 10])

plt.subplot(2, 1, 1)

plt.plot(sub1, sub2)

# plot 2:

sub3 = np.array([0, 1, 2, 3])

sub4 = np.array([10, 20, 30, 40])

plt.subplot(2, 1, 2)

plt.plot(sub3, sub4)
plt.show()
```

## Matplotlib bars


-> bar() function is used to draw bar graphs.

-> bar() takes the argument 'width' to set the width of the bars.


```python

plot1 = np.array(["A", "B", "C", "D"])

plot2 = np.array([3, 8, 1, 10])

plt.bar(plot1, plot2, color = "red", width = 0.1)

plt.show()
```


-> barh() - Used to display bars horizontally.

-> barh() - takes the argument 'height' to set the height of the bars.


```python

plt.barh(plot1, plot2, height = 0.1)
plt.show()
```

## Matplotlib Pie Chart


-> pie() function is used to draw pie chart.



```python


pie = np.array([35, 25, 25, 15])

plt.pie(pie)

plt.show()
```


-> As you can see the piec chart draws one piece (called a wedge) for each value in the array.

-> By default the plotting of the first wedge starts from the x-axis and moves counterclockwise.


### Labels 

-> To add labels in pie chart we use the parameter label


```python

pie = np.array([35, 25, 25, 15])

labels = np.array(["Apples", "Bananas", "Cherries", "Dates"])

plt.pie(pie, labels = labels)

plt.show()
```

### Start Angle


-> As mentioned the default start angle is at the x-axis, but you can change the start angle by specifying

a *startangle * parameter.

-> The *startangle* parameter is defined with an angle degree, default is 0

```python

pie = np.array([35, 25, 25, 15])

labels = np.array(["Apples", "Bananas", "Cherries", "Dates"])

plt.pie(pie, labels = labels, startangle = 90)

plt.show()
```


## Explode


-> Incase you may want one wedge to stand out, *explode* parameter can be employed.

-> The *explode* parameter, if specified, and not *None*, must be an array with one value for each wedge.


```python

pie = np.array([35, 25, 25, 15])

labels = np.array(["Apples", "Bananas", "Cherries", "Dates"])


explode = [0.2, 0, 0, 0]


plt.pie(pie, labels = labels, startangle = 90, explode = explode)



plt.show()
```


## Pic Shade

-> To add a shadow to the pie chart, *shadows* parameter is set to *True*


```python

pie = np.array([35, 25, 25, 15])

labels = np.array(["Apples", "Bananas", "Cherries", "Dates"])


explode = [0.2, 0, 0, 0]


plt.pie(pie, labels = labels, startangle = 90, explode = explode, shadow = True)



plt.show()
```
